```python
N = 1009*1013#112093
B = 1009 #Граница гладкости факторной базы
A = 10000 #Ширина диапозона поиска
```

```python
m = ceil(sqrt(N)); n(m)
```

```python
factor_base = [] # Массив, содержащий факторную базу
for prime in list(primes(2, B + 1)):
    if kronecker(prime,N)==1:#символы лежандра
        factor_base.append(prime)
factor_base
```

```python
D_table = dict()#таблица, с x=m+i ключём и q(x)=x^2-N значением. Содержит только гладкие числа
```

```python
#Формирование матрицы D
for i in range(m, m + A + 1):
    D_table[i] = (i^2 - N, [0]*len(factor_base))
```

```python
#Просеивание
D_table_sieved = D_table.copy()
for factor_index in range(0, len(factor_base)):
    factor = factor_base[factor_index]
    is_factor_complete = false
    while not is_factor_complete:
        is_factor_complete = true
        for key in list(D_table_sieved):
            if(D_table_sieved[key][0] % factor == 0):
                is_factor_complete = false
                D_table_sieved[key][1][factor_index] += 1
                D_table_sieved[key] = (D_table_sieved[key][0] / factor, D_table_sieved[key][1])
                if D_table_sieved[key][0]==1:
                    print(key)
                    print(D_table_sieved[key])
```

```python
# Объединям изначальную таблицу и просеянную. Оставляем только просеянные числа
D_table_result = dict()
len(D_table_sieved)
for key in list(D_table_sieved):
    if(D_table_sieved[key][0] == 1):
        D_table_result[key] = (D_table[key][0], D_table_sieved[key][1])
D_table_result
```

```python
#Построение матрицы A, у которой столбики - это векторы показателей полученных кладких чисел по модулю 2
matrix_array = []
for key in list(D_table_result):
    matrix_array.append(D_table_result[key][1])
m = matrix(matrix_array).transpose()
modded_m = sage.matrix.misc.matrix_integer_dense_rational_reconstruction(m.change_ring(ZZ), 2)
```

```python
# Нахождение решений СЛАУ
solutions = modded_m.right_kernel().basis()
solutions
gcd_result = 0
for solution in solutions:
    moded_vector = vector(Integers(2),list(solution))
    D_table_indexes = list(D_table_result)
    mult = 1
    x = 1
    for i in range(0,len(moded_vector)):
        if(moded_vector[i]==1):
            x *= D_table_indexes[i]
            mult *= D_table_result[D_table_indexes[i]][0]
    y = sqrt(mult)
    gcd_result = gcd(x-y, N)
    if gcd_result != 1 and gcd_result != x-y and gcd_result != N:
        break
    else:
        mult = 1
        x = 1
        
        
print(gcd_result)
if(gcd_result != 0):
    print(N/gcd_result)
```
